```python
import random
import numpy as np
import pandas as pd
def rn_num(rn):
    if rn>=0 and rn<=4:
        return "Head"
    elif rn>=5 and rn<=9:
        return "Tail"
profit=[]
gamecount=0
```

```python
def gambling():
  sno=[0 for i in range(100)]
  gamenum=[0 for i in range(100)]
  demand=[0 for i in range(100)]
  hcum=[0 for i in range(100)]
  tcum=[0 for i in range(100)]
  diff=[0 for i in range(100)]
  rn=[]
  end=0
  prof=0
  for i in range(0,100):
    gamenum[i]=""
    if  diff[i]<3:
      sno[i]=i+1
      rn.append(random.randint(0, 9))
      demand[i]=rn_num(rn[i])
      if i>0:
        if  demand[i]=="Head":
          hcum[i]=hcum[i-1]+1
          tcum[i]=tcum[i-1]

        else:
          tcum[i]=tcum[i-1]+1
          hcum[i]=hcum[i-1]

      if i==0:
        gamenum[i]=gamecount+1
        if  demand[i]=="Head":
          hcum[i]=hcum[i]+1
        else:
          tcum[i]=tcum[i]+1
      diff[i]=abs( tcum[i]- hcum[i])


    if  diff[i]==3:
      end=i
      profit=8-sno[i]
      gamenum[i]=""

      break
  df={"game number":gamenum[0:end+1],"s.no":sno[0:end+1],"rn":rn[0:end+1],"demand":demand[0:end+1],"demand":demand[0:end+1],"head cumlitve":hcum[0:end+1],"tail cumlitve":tcum[0:end+1],"diffrence ":diff[0:end+1]}
          
  df=pd.DataFrame(df)
  df.index = np.arange(1, len(df)+1)
  print(df)
  print("                         ")
  print("profit from this game is ",profit)
  return (profit,sno[i])
```

```python
totalprofit=[]
totaltry=[]
for i in range(5):
  x,y=gambling()
  gamecount=gamecount+1
  totalprofit.append(x)
  totaltry.append(y)
print("                         ")
print("total money paid",sum(totaltry))
print("                         ")
print("total profit",sum(totalprofit))
```
